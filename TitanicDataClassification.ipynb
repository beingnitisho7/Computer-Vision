```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data=pd.read_csv(r'''C:\Users\Dell\train.csv''')
data.head()
```

```python
data_age_mean = data['Age'].mean()
```

```python
data_fill = data['Age'].fillna(data_age_mean)
```

```python
X = data_fill
Y = data['Survived']

X_array = np.array(X).reshape(-1,1)
Y_array = np.array(Y).reshape(-1,1)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_array,Y_array)
```

```python
test_data=pd.read_csv(r'''C:\Users\Dell\test.csv''')
test_data
```

```python
test_data_age_mean = data['Age'].mean()
```

```python

test_data_avg = test_data['Age'].fillna(test_data_age_mean)
```

```python
X_test = test_data_avg
X_test_array = np.array(X_test).reshape(-1,1)
```

```python
new_survived_model = model.predict(X_test_array)
```

```python
model.score(X_array,Y_array)
```

```python
Survived = pd.DataFrame(new_survived_model,columns = ['Survived'])
Survived.head()
```

```python
PassengerId = test_data['PassengerId']
PassengerId.head()
```

```python
prediction_frame = pd.concat([PassengerId,Survived],axis = 1)
prediction_frame.head()
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_array)
print(Y_array)
print(y_pred)
confusion = confusion_matrix(Y_array, y_pred)
print("Confusion matrix:\n{}".format(confusion))
```
